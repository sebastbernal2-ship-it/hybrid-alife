Run a short experiment, then visualize fields and metrics.

```python
import sys, pathlib
sys.path.insert(0, str(pathlib.Path('..').resolve() / 'src'))

from hybrid_alife.experiments.runner import load_config, run_experiment
cfg = load_config('../configs/base.yaml')
state = run_experiment(cfg)
print('done', state.generation)
```

```python
from hybrid_alife.viz.plots import plot_world_fields, plot_agent_positions, plot_metrics
from pathlib import Path
run_dir = Path('../outputs/runs/smoke_hybrid_v1')
plot_world_fields(run_dir/'checkpoint_final.pkl', run_dir/'plots'/'world_fields.png')
plot_agent_positions(run_dir/'checkpoint_final.pkl', run_dir/'plots'/'agents.png')
plot_metrics(run_dir/'metrics.jsonl', run_dir/'plots'/'metrics.png')
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
for name in ('world_fields','agents','metrics'):
    plt.figure(figsize=(10,5))
    plt.imshow(mpimg.imread(run_dir/'plots'/f'{name}.png'))
    plt.axis('off'); plt.title(name); plt.show()
```
